```python
import matplotlib.pyplot as plt
```

```python
#VM 9
# Pie chart
labels = ['Not Used', 'Used']
sizes = [100-(674.605/(2925.99)*100), (674.605/(2925.99)*100)]
explode = (0, 0.1)
#add colors
plt.rcParams['font.size'] = 22
colors = ['seagreen','steelblue']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=270)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.title('Original Usage',fontsize = 20)
plt.savefig('images/vm106_orig_usage.png')


labels = ['Not Used', 'Used']
sizes = [100-(674.605/(819.998)*100), (674.605/(819.998)*100)]
explode = (0, 0.1)
#add colors
colors = ['seagreen','steelblue']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=0)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.title('My Model', fontsize = 20)
plt.savefig('images/vm106_mymod_usage.png')
plt.show()
```
